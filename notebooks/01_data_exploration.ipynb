```python
import torch
from torchvision import datasets, transforms
import matplotlib.pyplot as plt
import numpy as np
import os
```

```python
train_dir = "../data/TRAIN"
test_dir = "../data/TEST"

transform = transforms.Compose([
    transforms.Resize((128, 128)),
    transforms.ToTensor()
])


train_data = datasets.ImageFolder(train_dir, transform = transform)
test_data = datasets.ImageFolder(test_dir, transform = transform)

classes = train_data.classes
print("Classes:", classes)
```

```python
def imshow(img, title):
    img = img.numpy().transpose((1, 2, 0))
    plt.imshow(img)
    plt.title(title)
    plt.axis("off")
    plt.show()


for i in range(3):
    idx = np.random.randint(len(train_data))
    image, label = train_data[idx]
    imshow(image, f"Label:{classes[label]}")
```

```python
from collections import Counter
train_counts = Counter([label for _, label in train_data])
test_counts = Counter([label for _, label in test_data])
print("Training set distribution:", {train_data.classes[i] : count for i, count in train_counts.items()})
print("Testing set distribution:", {test_data.classes[i] : count for i, count in test_counts.items()})
```
